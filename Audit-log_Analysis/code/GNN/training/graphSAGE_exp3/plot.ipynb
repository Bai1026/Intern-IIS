```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('hidden_dim_2.csv')
df2 = pd.read_csv('hidden_dim_3.csv')
df3 = pd.read_csv('hidden_dim_4.csv')

df1 = df1[df1['support'] == 256000]
df2 = df2[df2['support'] == 256000]
df3 = df3[df3['support'] == 256000]
```

```python
import numpy as np

metric = 'f1-score'
labels = df1['label']

x = np.arange(len(labels))
plt.figure(figsize=(30, 12))

width = 0.4
plt.bar(x - width/2, df1[metric], width, label='hidden_dim = 256', color='b')
plt.bar(x + width/2, df2[metric], width, label='hidden_dim = 512', color='skyblue')

plt.xlabel('Labels')
plt.ylabel(metric.capitalize())
plt.title(f'Comparison of {metric.capitalize()} between two different hidden dimension')

plt.xticks(x, labels, rotation='vertical')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

metric = 'f1-score'
labels = df1['label']

x = np.arange(len(labels))

plt.figure(figsize=(20, 10))
plt.plot(x, df1[metric], marker='o', label='hidden_dim = 256', color='b')
plt.plot(x, df2[metric], marker='o', label='hidden_dim = 512', color='r')
plt.plot(x, df3[metric], marker='o', label='hidden_dim = 1024', color='g')

plt.xlabel('Labels')
plt.ylabel(metric.capitalize())
plt.title(f'Comparison of {metric.capitalize()} between three CSV files')

plt.xticks(x, labels, rotation='vertical')
plt.legend()
plt.show()
```
