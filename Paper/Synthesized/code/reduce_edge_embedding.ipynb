```python
OUTPUT_EMBEDDING_PATH = './'
MODEL_NAME = 'transE'
DIM = 50
EMBEDDING_NAME = f'{MODEL_NAME}_{DIM}.vec.json'
```

```python
import json
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

with open(f"{OUTPUT_EMBEDDING_PATH}/{EMBEDDING_NAME}", 'r') as file:
    data = json.load(file)
```

```python
emb_array = np.array(data['rel_embeddings.weight'])
```

```python
emb_array.shape
```

```python
scaler = StandardScaler()
emb_array_normalized = scaler.fit_transform(emb_array)

pca = PCA(n_components=16)
reduced_embeddings = pca.fit_transform(emb_array_normalized)

print(reduced_embeddings.shape)  # 应该是 (23, 2)
```
