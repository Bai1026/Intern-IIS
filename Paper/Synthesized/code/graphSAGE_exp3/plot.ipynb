# For plotting difference f1-score between different setting

```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('./csv_files/hidden_dim_64.csv') # dim=64
# df2 = pd.read_csv('./csv_files/hidden_dim_3.csv') # dim=512
df3 = pd.read_csv('./csv_files/hidden_dim_4.csv') # dim=1024
df4 = pd.read_csv('./csv_files/hidden_dim_7.csv') # dim=2048
df5 = pd.read_csv('./csv_files/transH_150_hidden_dim_1024.csv') # dim=2048
df6 = pd.read_csv('./csv_files/GCN_64.csv') # dim=64


df1 = df1[df1['support'] == 100]
# df2 = df2[df2['support'] == 256000]
df3 = df3[df3['support'] == 100]
df4 = df4[df4['support'] == 100]
df5 = df5[df5['support'] == 100]
df6 = df6[df6['support'] == 100]
```

```python
mapping_file = './new_mapping.txt'
label_mapping = {}
with open(mapping_file, 'r') as f:
    for line in f:
        parts = line.strip().split(': ')
        label_mapping[int(parts[1])] = parts[0]
```

```python
import numpy as np

metric = 'f1-score'
# labels = df1['label']
labels = df1['Unnamed: 0']

x = np.arange(len(labels))
plt.figure(figsize=(30, 12))

width = 0.4
plt.bar(x - width/2, df1[metric], width, label='hidden_dim = 64', color='b')
# plt.bar(x + width/2, df2[metric], width, label='hidden_dim = 512', color='skyblue')
plt.bar(x + width/2, df3[metric], width, label='hidden_dim = 1024', color='skyblue')

plt.xlabel('Labels')
plt.ylabel(metric.capitalize())
plt.title(f'Comparison of {metric.capitalize()} between two different hidden dimension')

plt.xticks(x, labels, rotation='vertical')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

metric = 'f1-score'
# labels = df1['label']
labels = df1['Unnamed: 0']

x = np.arange(len(labels))

plt.figure(figsize=(20, 10))
plt.plot(x, df1[metric], marker='o', label='hidden_dim = 64', color='b')
plt.plot(x, df3[metric], marker='o', label='hidden_dim = 1024', color='r')
plt.plot(x, df4[metric], marker='o', label='hidden_dim = 2048', color='g')

plt.xlabel('Labels')
plt.ylabel(metric.capitalize())
plt.title(f'Comparison of {metric.capitalize()} between three CSV files')

plt.xticks(x, labels, rotation='vertical')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter out the rows where the f1-score is equal to 1
df1_filtered = df1[df1[metric] != 1]
df3_filtered = df3[df3[metric] != 1]
df4_filtered = df4[df4[metric] != 1]

# Now, we'll update the labels and x-axis points accordingly.
labels_filtered = df1_filtered['Unnamed: 0']
x_filtered = np.arange(len(labels_filtered))

# Plotting
plt.figure(figsize=(20, 10))

plt.plot(x_filtered, df1_filtered[metric], marker='o', label='hidden_dim = 64', color='b')
plt.plot(x_filtered, df3_filtered[metric], marker='o', label='hidden_dim = 1024', color='r')
plt.plot(x_filtered, df4_filtered[metric], marker='o', label='hidden_dim = 2048', color='g')

plt.xlabel('Labels')
plt.ylabel(metric.capitalize())
plt.title(f'Comparison of {metric.capitalize()} between three CSV files')
plt.xticks(x_filtered, labels_filtered, rotation='vertical')
plt.legend()

# Adding grid
plt.grid(True)

plt.show()
```

```python
# Count the number of times df3 f1-score is greater than df4 f1-score
count_df3_greater_df4 = (df3_filtered[metric] > df4_filtered[metric]).sum()
print(f"Number of dim1024 f1-score is greater than dim2048 f1-score: {count_df3_greater_df4}")

# Count the number of times df4 f1-score is greater than df3 f1-score
count_df4_greater_df3 = (df4_filtered[metric] > df3_filtered[metric]).sum()
print(f"Number of dim2048 f1-score is greater than dim1024 f1-score: {count_df4_greater_df3}")

# Count the number of labels where df1 f1-score equals 0
count_df1_f1_score_zero = (df1[metric] == 0).sum()
print(f"Number of labels in dim64 where f1-score equals 0: {count_df1_f1_score_zero}")

# Total number of labels in df1
total_labels_df1 = len(df1)
print(f"Total number of labels of single triplet case: {total_labels_df1}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

metric = 'f1-score'
labels = df1['label']

x = np.arange(len(labels))

plt.figure(figsize=(20, 10))
plt.scatter(x, df1[metric], label='hidden_dim = 256', color='b')
plt.scatter(x, df2[metric], label='hidden_dim = 512', color='r')
plt.scatter(x, df3[metric], label='hidden_dim = 1024', color='g')

# 添加平均值的水平线
avg1 = df1[metric].mean()
avg2 = df2[metric].mean()
avg3 = df3[metric].mean()

# plt.axhline(y=avg1, color='b', linestyle='--', label='Avg. hidden_dim = 256')
plt.axhline(y=avg1, color='b', linestyle='--')
plt.axhline(y=avg2, color='r', linestyle='--')
plt.axhline(y=avg3, color='g', linestyle='--')

plt.xlabel('Labels')
plt.ylabel(metric.capitalize())
plt.title(f'Comparison of {metric.capitalize()} between three CSV files')
plt.xticks(x, labels, rotation='vertical')
plt.legend()

# 添加网格
plt.grid(True)

# 添加平均值的标签
plt.text(len(labels)+1, avg1, f'Avg. = {avg1:.2f}', color='b', va='center')
plt.text(len(labels)+1, avg2, f'Avg. = {avg2:.2f}', color='r', va='center')
plt.text(len(labels)+1, avg3, f'Avg. = {avg3:.2f}', color='g', va='center')

plt.show()
```
