```python
import json
from tqdm.notebook import tqdm

datasets = ['train', 'valid', 'test']

for dataset_name in tqdm(datasets):
    file_path = f"../data/exp3/training_data/secureBERT_50/{dataset_name}.jsonl"
    
    print(file_path)
    
    # 
    output_file_path = f"../data/exp3/training_data/secureBERT_50/{dataset_name}_updated.jsonl"
    with open(output_file_path, 'w') as output_file:
        
        #  JSONL 
        with open(file_path, 'r') as input_file:
            for line in tqdm(input_file):
                # 
                data = json.loads(line)
                
                # 'labels'
                data['labels'] = [int(label) for label in data['labels']]
                print(data['labels'])
                print(data)
                
                # 
                json.dump(data, output_file)
                output_file.write('\n')  # 
```
